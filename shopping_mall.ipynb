#  gender counting

```python
 sns.countplot(data=df,x='gender',hue='gender')
 plt.title('Shopping by Gender')
 plt.xlabel('Gender')
 plt.show()
```

# create age group

```python
min_age=df['age'].min()
max_age=df['age'].max()
bins=[18,25,32,39,46,53,60,67,74]
slot=['18-25','25-32','32-39','39-46','46-53','53-60','60-67','67-74']
df['age group']=pd.cut(df['age'],bins,labels=slot)
df
```

# shopping by age group and gender

```python
sns.countplot(data=df,x='age group',hue='gender')
plt.xlabel('Age Group')
plt.title('Shopping by Age Group')
plt.show()
```

# category wise sales value

```python
cat_value=df.groupby(['category','gender'])[['quantity','price']].sum()
cat_value=cat_value.reset_index()
sns.catplot(data=cat_value,x='category',y='price',kind='bar',height=2.75)
plt.xlabel('Category')
plt.ylabel('Total value')
plt.xticks(rotation=90)
plt.title('Category wise Sales value')
plt.show()
```

# Gender wise category sale

```python
sns.catplot(data=cat_value,x='category',y='price',hue='gender',kind='bar',height=2.75)
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Total value')
plt.title('Gender wise category sale')
plt.show()
```

# quantity wise category sale

```python
sns.catplot(data=cat_value,x='category',y='quantity',kind='bar',height=2.75)
plt.xlabel('Category')
plt.ylabel('Quantity')
plt.xticks(rotation=90)
plt.title('Quantity wise category sale')
plt.show()
```

#invoice_date column has object dtype

```python
df['invoice_date']=pd.to_datetime(df['invoice_date'],dayfirst=True)
```

```python
df.info()
```

#now invoice_date column is datetime dtpe

# creation of year column

```python
df['year']=df['invoice_date'].dt.year
df
```

 # creation of sale column

```python
df['sale_amt']=df['quantity']*df['price']
df
```

# mall wise yearly sale

```python
year_sale=df.groupby(['year','shopping_mall'])['sale_amt'].sum()
year_sale=year_sale.reset_index()
sns.catplot(data=year_sale,x='year',y='sale_amt',hue='shopping_mall',kind='bar')
plt.xlabel('Year')
plt.ylabel('Sales Value')
plt.title('Mall wise yearly sale')
plt.show()
```

 # contribution of shopping mall in total sale(3 years)

```python
shop_mall=df.groupby(['shopping_mall'])['sale_amt'].sum()
shop_mall=shop_mall.reset_index()
shopping_mall=shop_mall['shopping_mall'].tolist()
value=shop_mall['sale_amt'].tolist()
plt.pie(value,labels=shopping_mall,autopct='%1.2f%%')
plt.title('Contribution of Shopping mall')
plt.show()
```

# contribution of payment method

```python
sns.countplot(data=df,x='payment_method',hue='year')
plt.xlabel('Payment Method')
plt.ylabel('No. of People')
plt.title('Contribution of Payment Method')
plt.show()
```

# relation between age group and gender

```python
sns.relplot(data=df,x='category',y='age group',hue='gender',height=10)
plt.xlabel('Category')
plt.ylabel('Age Group')
plt.title('Relation between Category and Age Group with Gender')
plt.show()
```
